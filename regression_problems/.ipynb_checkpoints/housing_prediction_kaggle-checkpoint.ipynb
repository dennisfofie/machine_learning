```python
# handle all the letters
cat_results = []
num_results = []
for obj in x_train.columns:
    if x_train[obj].dtypes == 'object':
        cat_results.append(obj)
    else:
        num_results.append(obj)
        
        
```

```python
housing_cat = x_train[[result for result in cat_results]]
housing_cat.head()

housing_num = x_train[[num for num in num_results]]
housing_cat.head()

x_train['MSSubClass'].astype('int')
```

```python
# build tranformers for numbers columns
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
class FeatureDropper(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X.drop(['Alley', 'PoolQC', 'MasVnrType', 'FireplaceQu', 'MiscFeature','Fence'], axis=1, errors='ignore')
    
num_pipeline = Pipeline([('dropper',FeatureDropper()),("scaler", StandardScaler()), ('imputer', SimpleImputer(strategy='mean'))]
        )
num_result = num_pipeline.fit_transform(housing_num)
num_result
    
```

```python
# Build pipeline for cat features
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
full_pipeline = ColumnTransformer([("num_cols", num_pipeline, num_results), ("cat_cols", OneHotEncoder(), cat_results)])
```

```python
# model training
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(prepared_housing, y_train)

#evaluate it
from sklearn.metrics import mean_squared_error
pred_house = lr_model.predict(prepared_housing)
mse_score = mean_squared_error(y_train, pred_house)
rmse_score = np.sqrt(mse_score)
print(mse_score)
print()
print(rmse_score)
```

```python
from sklearn.tree import DecisionTreeRegressor
dr_model = DecisionTreeRegressor()
dr_model.fit(prepared_housing, y_train)

#evaluate it
pred_house_tree = dr_model.predict(prepared_housing)
tree_score_mse = mean_squared_error(y_train, pred_house_tree)
tree_score_rmse = np.sqrt(tree_score_mse)
print(tree_score_mse)
print()
print(tree_score_rmse)
```

```python
from sklearn.model_selection import cross_val_score
cross_score = cross_val_score(dr_model, prepared_housing, y_train, cv=10, scoring='neg_mean_squared_error')
tree_rmse = np.sqrt(-cross_score)
print(tree_rmse)
```

```python
def display(score):
    print('scores ', score)
    print('mean ', score.mean())
    print('std ', score.std())
    
display(tree_rmse)
```

```python
cross_score_lr = cross_val_score(lr_model, prepared_housing, y_train, cv=10, scoring='neg_mean_squared_error')
lr_model_rmse = np.sqrt(-cross_score_lr)
print(lr_model_rmse)
```
